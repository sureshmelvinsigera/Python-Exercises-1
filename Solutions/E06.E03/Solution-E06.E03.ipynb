# Solution: Exercise 6.E03

```python
# Initialize the number
number = 15
```

```python
# Case 1: Modulus of the division by two is zero
#         and the number is therefore even
if(number%2 == 0):
    print(str(number) + " is an even number.")
    
# Default: Modulus of the division by two is not zero
#          and the number is therefore odd
else:
    print(str(number) + " is an odd number.")
```
